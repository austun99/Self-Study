## Duplicates

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('stolen_vehicles(copy).csv', header=0)
```

```python
df.head()
```

There might be duplicate rows in our data set and we want to be able to remove them
- We can try using **df.drop_duplicates()**
    - This function will consider all the columns and checking if everything is the same, e.g including the index, which is not fair since all indexes are unique.

```python
df.drop_duplicates()
```

There is another way of using the same function, but this time we can even consider certain columns:
- df.drop_duplicates(subset=None, keep='first', inplace=True, ignore_index=True)
    - *subset=None* we can specify which columns to consider when making a duplicate decision e.g *subset=['vehicle_type','vehicle_desc']*
    - *keep='first'* means that it will keep the first occurance and delete the rest
    - *inplace=True* we will be making changes to the original data set
    - *ignore_index=True* means that we will be ignoring the index

```python
df.drop_duplicates(subset=None, keep='first', inplace=True, ignore_index=True)
```

```python
df
```

```python
#removing vehicles with the same 'vehicle_type' and 'make_id'
df.drop_duplicates(subset=['vehicle_type','make_id'], keep='first', inplace=True, ignore_index=True)
```
