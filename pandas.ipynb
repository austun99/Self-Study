```python
# Open file with pandas using a different encoding
df1 = pd.read_csv('Employee_Sample_Data.csv', header=0, encoding='latin-1')
df2 = pd.read_csv('Financials_Sample_Data.csv', header=0)
```

Data from Financials_Sample_Data.csv had currency in USD with dollar sign in front of each value
- This is why I had to first remove each of the $$$ signs in front of each number
- To do this I used the **df.replace** function
    - In this function I replaced '\$' with ''

```python
df1.replace(to_replace='\$', value='', regex=True, inplace=True)
df2.replace(to_replace='\$', value='', regex=True, inplace=True)
```

Data from Financials_Sample_Data.csv also had values in brackets "()" and numbers had spaces and comma's between them.
- This means that I had to get rid of these spaces, brackets and commas to make is usefull/workable

- To remove all the brackets before and after each value I used the **df.replace** function
    - I had to go column for column
    - In this function I replaced **r'\(([\d,]+)\)'** with **r'\1'**
        - r'\((\d+)\) is used to match and capture numbers inside parentheses. 
        - The (\d+) part captures one or more digits. 
        - The r'\1' in the replacement is a backreference to the captured digits, effectively removing the parentheses.

```python
#DATA CLEANING
#clean data to get rid of bracket around each number
#can be used to remove all types of stuff

# Specify the columns where you want to remove brackets
columns_to_clean = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']

# Remove brackets around numbers in the specified columns
for col in columns_to_clean:
    df2[col] = df2[col].replace(to_replace=r'\(([\d,]+)\)', value=r'\1', regex=True)

# Save the modified DataFrame back to a CSV file
df2.to_csv('Modified_Employee_Data.csv', index=False, encoding='utf-8')
```

The same concept for removing dollar signs can be used to remove the commas and spaces between the numbers:
Data from Financials_Sample_Data.csv also had values in brackets "()" and numbers had spaces and comma's between them.
- To remove all spaces and commas of each value I used the **df.replace** function
    - I had to go column for column
    - In this function I replaced **[',', ' ']** with **''**
    - And just convert each column to and integer

```python
#DATA CLEANING
#removing the spaces and commas between numbers to make them more workable with in python

# Remove spaces and commas from numbers in the specified columns
for col in columns_to_clean:
    df2[col] = df2[col].replace(to_replace=[',', ' '], value='', regex=True)

# Convert the columns to integers
df2[columns_to_clean] = df2[columns_to_clean].astype(int)

# Save the modified DataFrame back to a CSV file
df2.to_csv('Modified_Employee_Data.csv', index=False, encoding='utf-8')
```
